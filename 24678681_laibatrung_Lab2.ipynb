## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np
arr=np.random.randint(0,1000, size =100)
print(arr[:10])
print(arr[-5:])
result = arr[(arr % 5 == 0) & (arr > 100)]
print(result)
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
import numpy as np
arr=np.random.randint(0,1000, size =100)
result = arr[(arr < 100)]
print(len(result))
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
import numpy as np
arr=np.random.randint(0,1000, size =100)
Sum = np.sum(arr)
print("Tong:", Sum)
Ave = np.mean(arr)
print("TB:", Ave)
TrungVi = np.median(arr)
print("trungvi:", TrungVi)
Q1 = np.percentile(arr, 25)
Q2 = np.percentile(arr, 50)   
Q3 = np.percentile(arr, 75)
print("Q1:", Q1)
print("Q2:", Q2)
print("Q3:", Q3)
p20 = np.percentile(arr, 20)
print("PhanVi 20:", p20)
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
import numpy as np
arr = np.random.randint(0, 1001, size=100)
for i in range(1, 6):
    sample = np.random.choice(arr, 10, replace=False)
    print(f"Lần {i}:", sample)
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
values, counts = np.unique(arr, return_counts=True)

nhieunhat = values[np.argmax(counts)]
solan = counts.max()

print("Phần tử xuất hiện nhiều nhất:", nhieunhat)
print("Số lần xuất hiện:", solan)
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, linestyle='--', label='cos(x)')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Đồ thị hàm sin(x) và cos(x)")
plt.grid()
plt.legend()

plt.show()
```

### Câu 5: Vẽ các đồ thị con

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2*np.pi, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))


ax1.plot(x, y_sin, 'b-', linewidth=2, label='sin(x)')
ax1.set_ylim(-1.2, 1.2)
ax1.set_xlim(0, 2*np.pi)
ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)
ax1.grid(True, alpha=0.3)
ax1.legend(loc='upper right')
ax1.set_title('Đồ thị hàm sin(x) và cos(x)', fontsize=14, fontweight='bold')


ax2.plot(x, y_cos, 'r-', linewidth=2, label='cos(x)')
ax2.set_ylim(-1.2, 1.2)
ax2.set_xlim(0, 2*np.pi)
ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.axvline(x=0, color='k', linewidth=0.5)
ax2.grid(True, alpha=0.3)
ax2.legend(loc='upper right')
ax2.set_xlabel('x', fontsize=12)


plt.tight_layout()
plt.show()
```

### Câu 8: Vẽ histogram

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt


data = np.random.normal(size=10000)


plt.figure(figsize=(8, 6))
plt.hist(data, bins=50, color='Blue', alpha=0.9)


plt.title('Histogram', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)


plt.grid(axis='y', alpha=0.3)

# Lưu hình
plt.tight_layout()

# Hiển thị
plt.show()
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt


data = np.random.normal(size=10000)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), 
                                gridspec_kw={'height_ratios': [3, 1]})

ax1.hist(data, bins=50, color='mediumpurple', edgecolor='black', 
         alpha=0.8, density=True)
ax1.set_title('Đồ thị hàm sin(x) và cos(x)\nHistogram', fontsize=12)
ax1.set_ylabel('Mật độ', fontsize=10)
ax1.set_xlim(-4, 5)
ax2.boxplot(data, vert=False, widths=0.5)
ax2.set_xlabel('Boxplot', fontsize=10)
ax2.set_xlim(-4, 5)
ax2.set_yticks([])

plt.tight_layout()

plt.show()

```
